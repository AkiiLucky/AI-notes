```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data.txt',sep=' ')
data
```

```python
def H(p):   #计算经验熵
    p = np.array(p)
    p = p[p != 0]
    p = p / np.sum(p)
    return np.sum(-p*np.log2(p))
```

```python
def H_D_A(data,A): #经验条件熵
    a = []
    for i in data[A]:
        if i not in a:
            a.append(i)
    print(A)
    sum = 0
    for i in a:
        p = []
        data_i = data[data[A]==i]
        p.append(len(data_i[data_i['类别']=='同意']))
        p.append(len(data_i[data_i['类别']=='不同意']))
        sum += H(p)*(len(data_i)/len(data))
    return sum
```

```python
def g(data,A):  #信息增益
    a = []
    for i in data['类别']:
        if i not in a:
            a.append(i)
    p = []
    for i in a:
        p.append(len(data[data['类别']==i]))
    return H(p) - H_D_A(data,A)

def gr(data,A): #信息增益比
    a = []
    for i in data[A]:
        if i not in a:
            a.append(i)
    p = []
    for i in a:
        p.append(len(data[data[A]==i]))
    return g(data,A)/H(p)
```

```python
print(g(data,'年龄'))
print(g(data,'有工作'))
print(g(data,'有房子'))
print(g(data,'信用'))
```

```python
print(gr(data,'年龄'))
print(gr(data,'有工作'))
print(gr(data,'有房子'))
print(gr(data,'信用'))
```

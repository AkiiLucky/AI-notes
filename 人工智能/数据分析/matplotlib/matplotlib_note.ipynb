```python
import matplotlib.pyplot as plt
import math
```

```python
a = [x * 3.14 / 5 for x in range(0, 100)]
plt.plot(a)
plt.show()
```

```python
b = [math.sin(x)*3 + x for x in a]
plt.plot(b)
plt.show()
```

```python
fig = plt.figure()
fig, ax = plt.subplots()  # 创建空图表
```

```python
fig = plt.figure()
fig, ax = plt.subplots(2, 3)
```

```python
fig = plt.figure()
fig, ax = plt.subplots()

plt.rcParams['font.sans-serif'] = ['SimHei']  # 修改图形的字体，以支持中文

ax.set_title("测试")  # 设置图表标题

ax.set_xlim(0, 15)   # 设置横坐标范围 
ax.set_xticks([1, 3, 5, 7, 11])  # 设置横轴的刻度值
ax.set_xticklabels(['one', 'three', 'five', 'seven', 'eleven'])  # 设置横轴的刻度值的标签
ax.set_xlabel('x轴')  # 设置坐标轴名称

# ax.grid(True) # 显示网格线
ax.grid(True, which='major', axis='x') # 显示垂直x轴的网格线

ax.plot(a)
ax.plot(b)

ax.legend(['a', '3 * sin(a) + a'])  # 显示图形标注信息

# fig.savefig('fig01.png')
```

```python
a = range(10)
b = [x*x for x in a]
fig = plt.figure()
fig, ax = plt.subplots()
ax.scatter(a,b)  # 散点图
```
